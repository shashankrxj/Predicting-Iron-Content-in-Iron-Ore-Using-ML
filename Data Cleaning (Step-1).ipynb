## Stage #1: From the earlier analysis, there are no missing values and the data is relatively clean. However, the columns have varying scales and might need normalization.

```python
df=df.set_index('date')
```

# Step 3: Plot the data distribution to determine the presence of any outliers

```python
import matplotlib.pyplot as plt
from matplotlib import style

# Apply 'ggplot' style to the plots
plt.style.use('ggplot')

# Create histograms for the dataframe with specified bins and figure size
df.hist(bins=70, figsize=(17, 15))

# Display the plot
plt.show()

# Set the title of the figure with white color
plt.suptitle('Figure Title', color='w')
```

```python
df.head()
```

# Step 4: Eliminate outliers using the z-score method by removing data points beyond 2 standard deviations

```python
df_copy=df
```

```python
# Filter out data points that fall outside 2 standard deviations from the mean

from scipy import stats

# Create a copy of the dataframe excluding outliers based on z-score
df_copy = df[(np.abs(stats.zscore(df)) < 2).all(axis=1)]
df_copy.info()

# Explanation:
# The z-score for each value is calculated relative to the column mean and standard deviation.
# The absolute value of the z-score is used to focus on whether the z-score magnitude is below the threshold, ignoring direction.
# The condition (all(axis=1)) ensures that all columns in each row meet the constraint.
# The resulting boolean mask is used to filter the original dataframe.
```

```python
df_copy.head()
```

```python
# Apply 'ggplot' style to the plots
plt.style.use('ggplot')

# Create histograms for the filtered dataframe with specified bins and figure size
df_copy.hist(bins=70, figsize=(17, 15))

# Display the plot
plt.show()

# Set the title of the figure with a specific color
plt.suptitle('Figure Title', color='black')

# Display information about the filtered dataframe
df_copy.info()
```

# Step 5: Examine the data to determine if resampling is necessary

```python
#Seeing impact of resampling using only first sample of every hour on the distribution
# Apply 'ggplot' style to the plots
plt.style.use('ggplot')

# Analyze the effect of resampling by taking the first sample of every hour and examining the distribution
df_first = df_copy.resample('H').first()
df_first.hist(bins=70, figsize=(17, 15))

# Display the plot
plt.show()

# Set the title of the figure with a specific color
plt.suptitle('Figure Title', color='black')

# Display information about the resampled dataframe
df_first.info()
```

```python
#Seeing impact of resampling using hourly mean on the distribution

# Apply 'ggplot' style to the plots
plt.style.use('ggplot')

# Evaluate the effect of resampling by calculating the hourly mean and examining the distribution
df_mean = df_copy.resample('H').mean()
df_mean.hist(bins=70, figsize=(17, 15))

# Display the plot
plt.show()

# Set the title of the figure with a specific color
plt.suptitle('Figure Title', color='black')

# Display information about the resampled dataframe
df_mean.info()
```

## Stage 2: It was decided to retain all data excluding outliers. Resampling significantly decreases the data available for modeling (by a factor of 180!). After outlier removal, approximately 346K data points remain.

# Step 6: Features Extraction 

```python
# Removing the '% Iron Concentrate' column as it is a result of laboratory analysis and closely correlated with '% Silica'
df_copy.drop(columns=['% Iron Concentrate'], inplace=True)
df_copy.head()
```

```python
ml_mining_data=df_copy
```

```python
#Plot a heatmap to review correlations
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
p = sns.heatmap(ml_mining_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
```

```python
# MANUALLY SELECTING FEATURES

df = ml_mining_data.drop(['% Silica Feed', 'Starch Flow', 'Ore Pulp Flow', 'Flotation Column 01 Air Flow',
                          'Flotation Column 03 Air Flow', 'Flotation Column 04 Air Flow', 'Flotation Column 05 Air Flow',
                          'Flotation Column 01 Level', 'Flotation Column 02 Level', 'Flotation Column 03 Level', 
                          'Flotation Column 05 Level', 'Flotation Column 06 Level'], axis=1)

# Criteria for selection:
# 1) Correlation of features with the target variable (note: some features with low correlation were retained due to criterion #3)
# 2) Excluding collinear data points to avoid overfitting
# 3) Analyzing feature distribution to ensure continuous data (note: some features with poor distribution were included because of criterion #1)
        
# Below are the columns retained based on manual decisions:
# ml_mining_data['% Iron Feed', 'Amina Flow', 'Ore Pulp pH', 'Flotation Column 02 Air Flow',
#                 'Flotation Column 06 Air Flow', 'Flotation Column 07 Air Flow',
#                 'Flotation Column 04 Level', 'Flotation Column 07 Level',
#                 '% Silica Concentrate']
```

```python
# FEATURE ENGINEERING
# Combine the remaining air flow features by calculating their mean, and do the same for the level data. This approach is based on the similarity of the data.
# Reducing the number of features will simplify the machine learning model-building process.

df.head()

# Create new columns representing the averages of the specified features
df['avg_air_flow_267'] = df[['Flotation Column 02 Air Flow', 'Flotation Column 06 Air Flow', 'Flotation Column 07 Air Flow']].mean(axis=1)
df['avg_float_level_47'] = df[['Flotation Column 04 Level', 'Flotation Column 07 Level']].mean(axis=1)
df.head()

# Plot the correlation heatmap to review feature relationships
plt.figure(figsize=(20, 10))
p = sns.heatmap(df.corr(), annot=True)
```

```python
#Further cleaning of dataset and final view of correlations

df_ml=df.drop(['Flotation Column 02 Air Flow','Flotation Column 06 Air Flow','Flotation Column 07 Air Flow','Flotation Column 04 Level', 'Flotation Column 07 Level'], axis = 1) 
df_ml.head()
cols = df_ml.columns.tolist()
cols
cols = cols[-1:] + cols[:-1]
cols = cols[-1:] + cols[:-1]
df_ml = df_ml[cols]
df_ml.head()
df_ml.info()
plt.figure(figsize=(20, 8))
p = sns.heatmap(df_ml.corr(), annot=True);

df_ml.hist(bins = 70, figsize = (10,10))
plt.show()
plt.suptitle('figure title', color='w')
```
